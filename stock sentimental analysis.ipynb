```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Prasadumarji\Downloads\Combined_News_DJIA.csv.zip", encoding ="ISO-8859-1")
df.head()
```

```python
train = df[df['Date']< '20150101']
test = df[df['Date']> '20141231']
```

```python
#### Removing Punctuation marks
data = train.iloc[:,2:27]
data.replace("[\]",regex=True, inplace=True)

#### Replacing column name for ease of access
lst1 = [i for i in range(25)]
index_name = [str(i) for i in lst1]
data.columns = index_name
data.head()
```

```python
###3 Converting headline to lower case
for index in index_name:
    data[index]= data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[2]
```

```python
#### CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
### Implement bag of words
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
traindataset[0]
```

```python
#### implement randomforestclassifier
randomclassifier = RandomForestClassifier(n_estimators =200, criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
import numpy as np
```

```python
#### test transform
test_transform = []
for row in range(0,len(data.index)):
    test_transform.append(' '.join(str(x) for x in data.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
test.loc[1611,:]
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(test['Label'],predictions)
print(matrix)
report = classification_report(test['Label'],predictions)
print(report)
score = accuracy_score(test['Label'],predictions)
print(score)
```
